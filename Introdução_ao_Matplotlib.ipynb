**Gráfico de barras aninhadas**

```python
# Criar variaveis para o grafico de barras

# Quantidade de vendas para o Produto A
valores_produto_A = [6,7,8,4,4]
# Quantidade de vendas para o Produto B
valores_produto_B = [3,12,3,4.1,6]

# Cria eixo x (indice) para produto A e produto B com uma separação de 0.25 entre as barras
index = np.arange(len(valores_produto_A))
bar_width = 0.25


# Plota as barras
plt.bar(index, valores_produto_A, width=bar_width, label='Produto A', color='b', align='center')
plt.bar(index + bar_width, valores_produto_B, width=bar_width, label='Produto B', color='y')

# coloca o nome dos meses como label do eixo x
meses = ['Agosto','Setembro','Outubro','Novembro','Dezembro']
plt.xticks(index + bar_width/2, meses)
# plt.xticks(np.arange(0, 5), meses)

# insere uma legenda no gráfico
plt.legend()

plt.title("Quantidade de Vendas")
plt.show()
```

### Customizando gráfico de pizza

* Adicionando informações de percentual, sombra, separação das partes e legenda

```python
vendas = [3000, 2300, 1000, 500]
labels = ['E-commerce', 'Loja Física', 'e-mail', 'Marketplace']

# define o nível de separabilidade entre as partes, ordem do vetor representa as partes
explode = (0.1, 0, 0, 0) 

# define o formato de visualização com saída em 1.1%%, sombras e a separação entre as partes
plt.pie(vendas, labels=labels, autopct='%.1f%%', shadow=True, explode=explode)

# inseri a legenda e a localização da legenda.
plt.legend(labels, loc=3)


# define que o gráfico será plotado em circulo
plt.axis('equal')

plt.show()
```

### Grafico com dois eixos

**Vamos Calcular a Precipitataçao em escala anual (acumulado) e a Temperatura Média do ar em escala anual**

```python
precip_anual = df.Precipitation.resample('1Y').sum()
temp_anual = df.Mean_Temperature.resample('1Y').mean()
```

```python
# Criando Figura
fig, ax = plt.subplots(figsize=(14,5))
# Plot
ax.bar(precip_anual.index, precip_anual, 
       width=150, 
       color='royalblue',
       edgecolor='black',
       label='Precipitation')

# Definir labels
ax.set_ylabel('Precipitation [mm]', fontdict=dict(fontweight='bold', fontsize=12))
ax.set_xlabel('Time', fontdict=dict(fontweight='bold', fontsize=12))
ax.set_title('Curso de Python', fontdict={'fontweight':'bold', 'fontsize':14})

# Adicionar grids
ax.grid(alpha=0.3)

# Outro eixo
ax1 = ax.twinx()

# PLot da precipitacao
ax1.plot(temp_anual, marker='o', color='tab:red', label='Mean Temperature')
ax1.set_ylabel('Mean Temperature [$^\circ$C]', 
               fontdict=dict(fontweight='bold', fontsize=12), 
               rotation=270, labelpad=20)


# legendas
h1, l1 = ax.get_legend_handles_labels()
h2, l2 = ax1.get_legend_handles_labels()
ax1.legend(h1+h2, l1+l2, loc='upper right', fontsize=10, frameon=False)
```

* **Outro exemplo, utilizando subplots**


Podemos guardar os valores das variaveis em objetos para usar nos métodos de plotagem do matplotlib

**Exemplo**

```python
max_temperature = df['Max_Temperature']
mean_temperature = df['Mean_Temperature']
min_temperature = df['Min_Temperature']
precipitation = df['Precipitation']
```

```python
# Parametros
nrows = 2
ncols = 2

# Criando Figura
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(figsize=(15,7), 
                                             nrows=nrows, ncols=ncols, 
                                             sharex=False, sharey=False,
                                            gridspec_kw={'wspace':0.2, 'hspace':0.2})
ax1.plot(precipitation, color='royalblue')
# Definir labels
ax1.set_ylabel('Precipitação')
ax1.set_xlabel('Tempo')
ax1.annotate("(A)", xy=(0.05, 0.9), xycoords="axes fraction", fontsize=12, fontweight='bold')
ax2.plot(max_temperature, color='firebrick')
ax2.set_ylabel('Temperatura Máxima')
ax2.set_xlabel('Tempo')
ax2.annotate("(B)", xy=(0.05, 0.9), xycoords="axes fraction", fontsize=12, fontweight='bold')
ax3.plot(mean_temperature, color='tab:orange')
ax3.set_ylabel('Temperatura Média')
ax3.set_xlabel('Tempo')
ax3.annotate("(C)", xy=(0.05, 0.9), xycoords="axes fraction", fontsize=12, fontweight='bold')
ax4.plot(min_temperature, color='indigo')
ax4.set_ylabel('Temperatura Mínima')
ax4.set_xlabel('Tempo')
ax4.annotate("(D)", xy=(0.05, 0.9), xycoords="axes fraction", fontsize=12, fontweight='bold')

# Titulo geral
fig.suptitle('Gráficos no Matplotlib', y=0.95, fontsize=18, fontweight='bold')
```

* **Exemplo com numero de plots impar**

```python
# Parametros
nrows = 2
ncols = 2

# Criando Figura
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(figsize=(15,7), 
                                             nrows=nrows, ncols=ncols, 
                                             sharex=False, sharey=False,
                                            gridspec_kw={'wspace':0.2, 'hspace':0.2})

plt.delaxes(ax4)


ax1.plot(precipitation, color='royalblue')
# Definir labels
ax1.set_ylabel('Precipitação')
ax1.set_xlabel('Tempo')
ax1.annotate("(A)", xy=(0.05, 0.9), xycoords="axes fraction", fontsize=12, fontweight='bold')
ax2.plot(max_temperature, color='firebrick')
ax2.set_ylabel('Temperatura Máxima')
ax2.set_xlabel('Tempo')
ax2.annotate("(B)", xy=(0.05, 0.9), xycoords="axes fraction", fontsize=12, fontweight='bold')
ax3.plot(mean_temperature, color='tab:orange')
ax3.set_ylabel('Temperatura Média')
ax3.set_xlabel('Tempo')
ax3.annotate("(C)", xy=(0.05, 0.9), xycoords="axes fraction", fontsize=12, fontweight='bold')

# Titulo geral
fig.suptitle('Gráficos no Matplotlib', y=0.95, fontsize=18, fontweight='bold')
```

## Scatter plots

O gráfico de dispersão é um gráfico de dois conjuntos de dados ao longo dos dois eixos. É usado para visualizar a relação entre as duas variáveis.

Se o valor ao longo do eixo Y parece aumentar à medida que o eixo X aumenta (ou diminui), isso pode indicar uma relação linear positiva (ou negativa). Considerando que, se os pontos forem distribuídos aleatoriamente sem padrão óbvio, isso pode indicar uma falta de relacionamento dependente.

Em python matplotlib, o gráfico de dispersão pode ser criado usando **`plt.scatter`** ou **`ax.scatter`** . Usando essas funções, você pode adicionar mais recursos ao seu gráfico de dispersão, como alterar o tamanho, a cor ou a forma dos pontos


**Exemplo**


* **Grafico de dispersao entre a Temperatura Máxima e Temperatura Mínima**

```python
fig, ax = plt.subplots()
ax.scatter(max_temperature, min_temperature, marker='+')
ax.set(xlabel='Temperatura Máxima ($^\circ$C)', ylabel='Temperatura Mínima ($^\circ$C)', title='Gráfico Dispersão')
```

* **Grafico de dispersao mais linha de regressão** 

Primeiro vamos calcular a regressão linear utilizando o numpy para prosseguir com o plot

```python
# Guardando os valores em variaveis 
X = df.Precipitation
Y = df.Min_Temperature
```

```python
# Calculo Regressão Linear
a, b = np.polyfit(X, Y, 1)
```

```python
fig, ax = plt.subplots(figsize=(12,8))
ax.scatter(X, Y, marker='.', color='k')
ax.plot(X, a*X + b, color='red', linestyle='-')  # Regression line
ax.set(xlabel='Precipitation (mm)', ylabel='Min Temperature ($^\circ$C)', title='Scatter Plot')
ax.minorticks_on()
```

## Grafico de Preenchimento


**Tipo de grafico utilizado quando queremos analisar desvios, maximos e mínimos, percentis entre outros tipos de analises**

```python
fig, ax = plt.subplots(figsize=(8,7))
ax.plot(mean_temperature, color='black', label='Mean Temperature')
ax.fill_between(df.index, min_temperature, max_temperature, color='royalblue', alpha=0.3, label='Max-Min Temperature')
ax.legend(frameon=False, loc='upper left')
```
